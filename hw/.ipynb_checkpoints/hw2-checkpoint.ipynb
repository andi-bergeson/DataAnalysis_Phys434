```python
%matplotlib inline
import numpy as np
import math
from scipy import stats, special, signal
from matplotlib import pyplot as plt
```

# Problem 1

### "Which is more probable when rolling 2 six-sided dice: rolling snake eyes (two 1's), or rolling 7's (dice sum to seven)? What is the ratio of the probabilities?"

```python
dice = stats.randint(1,7)

snakeeyes = 100*(dice.pmf(1) * dice.pmf(1))
sevens = 100*(2*(dice.pmf(1)*dice.pmf(6)) + 2*(dice.pmf(2)*dice.pmf(5)) + 2*(dice.pmf(3)*dice.pmf(4)))

ratio = sevens/snakeeyes

print("Probability of rolling snake eyes : {0:.3f}% \nProbability of rolling seven's : {1:.3f}% \nRolling seven's is {2:g} times more likely".format(
    snakeeyes, sevens, ratio))
```

# Problem 2

### "Following what we did in class, show how to use the convolution operator to determine the probability of the sum of 2 six sided dice. Do both analytically (math & counting) and numerically (computer program). Beware the implicit definition of the values (x-axis on histogram)."

```python
fig, ax = plt.subplots(1,1)
fig.set_size_inches(16,6)

v = np.linspace(1,6,6)
conv = signal.fftconvolve(dice.pmf(v),dice.pmf(v),mode='full')

v_new = np.linspace(2,12,11)
ax.scatter(v_new,conv,marker='s')

ax.set_title("Probability of the sum of two 6-sided dice, using convolution")
ax.set_xlabel("Sum of Two Dice")
ax.set_ylabel("Probability")
plt.grid()
```

## Analytically, convolution of two discrete functions looks like $\displaystyle\sum_{n=2}^{12} \displaystyle\sum_{i=1}^{6} P(i)P(n-i)$ where $n$ is the array of all possible values for the sum of two dice, and $P(i)$ always equals 1/6.

# Problem 3

### "Calculate the mean and the variance of the distribution in problem 2. Hint: this is surprisingly tricky, make sure your result makes sense."

```python
mean = 0.
for i in range(len(conv)):
    adding = conv[i]*v_new[i]
    mean += adding
mean
```

```python
exp = 0.
for i in range(len(conv)):
    add = conv[i]*(v_new[i]**2)
    exp += add
exp
```

```python
variance = np.sqrt(exp - mean**2)
```

```python
print("The mean of my convolved distribution is {0:g}, and the variance is {1:.3f}.".format(mean,variance))
```

# Problem 4

### "Repeat 2, and graph the average of 10 dice. Is this is a Gaussian distribution? Explain in depth."

```python
fig, ax = plt.subplots(1,1)
fig.set_size_inches(18,7) 

z = np.linspace(1,6,6)
convten = dice.pmf(z)

rolls = 10
for i in range(rolls):
    convten = signal.fftconvolve(convten,dice.pmf(z),mode='full')

z_new = np.linspace(rolls*2,rolls*12,len(convten))
ax.scatter(z_new,convten,marker='s',label="Convolved Discrete PDF")

#now to fit a gaussian to our data... must find mean/variance

meanten = 0.
for i in range(len(convten)):
    adding = convten[i]*z_new[i]
    meanten += adding

expten = 0.
for i in range(len(convten)):
    add = convten[i]*(z_new[i]**2)
    expten += add
    
varten = np.sqrt(expten - meanten**2)

x = np.linspace(20,120,101)
ax.plot(x,stats.norm.pdf(x,loc=meanten,scale=varten),label="Gaussian, Unchanged")
ax.plot(x,2*stats.norm.pdf(x,loc=meanten,scale=varten),label="Gaussian, Multiplied by 2")

ax.set_title("Probability of the sum of ten 6-sided dice")
ax.set_xlabel("Sum of Ten Dice")
ax.set_ylabel("Probability")

plt.grid()
plt.legend()
plt.show()
```

## By overplotting a Gaussian, we can see that the convolved result of the PDF's of ten dice rolled together yields a larger peak than that of a Gaussian. This means that, as the PDF's were continuing to be multiplied together, the higher proportion of probability in the center resuled in a peak that continually extends past that of a Gaussian. If we were to measure, say, the sum of 7 dice, the PDF might moreso resemble that of a Gaussian... however, the weird sawtooth shape of the original PDF definitely smoothed into a Gaussian-esque shape.

# Problem 5

### "Show that the sum and average of an initially Gaussian distribution is also a Gaussian (can be analytic or numerical). How does the standard deviation of the resulting sum or average Gaussian change? This is a hugely important result. Explore what this means for integrating a signal over time."

```python
fig, ax = plt.subplots(1,1)
fig.set_size_inches(16,6)

b = np.linspace(-3,3,100)
gau = stats.norm.pdf(b)

ax.plot(b,gau,label='Normal Gaussian')

summed = gau+gau

m = 0.
for i in range(len(summed)):
    adding = summed[i]*b[i]
    m += adding
root = 0.
for i in range(len(summed)):
    add = (summed[i]-m)**2
    root += add
sd = np.sqrt(root/len(summed))

print("\n The average of a Gaussian when summed did not shift from the original mean, but the standard deviation decreased from 1 to {0:.5f}.\n".format(sd))

ax.plot(b,summed,label="Sum of Two Gaussians")

plt.legend()
plt.show()
```

## As a Gaussian distribution is summed, the standard deviation decreases and the distribution narrows. It loses properties that identify it as a Gaussian (i.e., the area under the curve no longer confines itself to the specified sigma) but will remain at the mean and become more accurate. 

## This is significant, because it implies that the more measurements we make of a phenomenon pulled from a Gaussian, the more the peak narrows and the more we are certain of the true mean.
